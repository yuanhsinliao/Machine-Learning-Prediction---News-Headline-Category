```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Dataset /News Category (Text).csv")
```

```python
X = df["headline"]
Y = df["category"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X_cv = CountVectorizer().fit_transform(X)
```

```python
X_cv.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_cv, Y)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LogisticRegression()
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test,pred)
print(cm)
print("accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
from sklearn import ensemble
```

```python
model = ensemble.RandomForestClassifier()
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test,pred)
print(cm)
print("accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_tfidf = TfidfVectorizer().fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_tfidf, Y)
```

```python
model = linear_model.LogisticRegression()
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test,pred)
print(cm)
print("accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
model = ensemble.RandomForestClassifier()
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test,pred)
print(cm)
print("accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```
